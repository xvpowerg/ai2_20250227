    * Bryant, P. G. 和 Smith, M (1995) Practical Data Analysis: Case Studies in Business Statistics. Homewood, IL: Richard D. Irwin Publishing:

# Conditional Filtering

```python
import pandas as pd
import numpy as np
df = pd.read_csv('tips.csv')
```

```python
df.head()
```

```python
bool_series = df["total_bill"] > 30
df[bool_series].head()
```

```python
df[df["sex"]== "Male"].head()
```

```python
df[ (df["total_bill"] > 30) & (df["sex"] == "Male") ]
```

```python
df[(df["day"] == "Sun") | (df["day"] == "Sat")]
```

```python
options = ["Sat","Sun"]
df[df["day"].isin(options)]
```

# Rolling

```python
import pandas as pd
df = pd.read_csv("starbucks.csv",index_col="Date",parse_dates=True)
df.index
```

```python
df.head()
```

```python
df["Close"].plot(figsize=(12,5))
```

```python
newDf = df.rolling(window=7).mean().head(15)
newDf = newDf.bfill()
newDf
```

```python
newDf.iloc[0:7] = 0
newDf
```

```python
df["Close"].plot(figsize=(12,5))
df.rolling(window=5).mean().bfill()["Close"].plot(figsize=(12,5))
df.rolling(window=15).mean().bfill()["Close"].plot(figsize=(12,5))
df.rolling(window=30).mean().bfill()["Close"].plot(figsize=(12,5))
```
