```python
import pandas as pd
import numpy as np
df = pd.read_csv("tips.csv")
```

```python
df.head()
```

```python
def yelp(price):
    if price < 10:
        return "$"
    elif price >=10 and price < 30:
        return "$$"
    else:
        return "$$$"
```

```python
df["total_bill"].mean()
```

```python
df["Expensive"] = df["total_bill"].apply(yelp)
df.head(20)
```

```python
df["total_bill"].apply(lambda bill:bill * 0.18)
```

```python
df.head()
```

```python
def quality(total_bill,tip):
    if tip / total_bill > 0.25:
        return "Generous"
    else:
        return "Other"
```

```python
df["Tip Quality"] = df[["total_bill","tip"]].apply(lambda df:quality(df["total_bill"],df["tip"]),axis=1)
df.head()
```

# np.vectorize(quality) 會產生一個函數 最終回傳一個向量

```python
import numpy as np
myFunc = np.vectorize(quality)    
df["Tip Quality np"] = myFunc(df["total_bill"],df["tip"])
df.head()
```
