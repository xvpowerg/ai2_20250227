# 類別圖表 - 類別內的統計估計
在資料分析中，我們經常會遇到類別資料，這些資料是以不同的組別（例如國家或公司）來分類的。不同於連續資料（例如年齡或價格），類別資料之間不存在"中間值"。例如，"美

國"和"法國"之間沒有一個國家值，"Google"和"Apple"之間也沒有一個公司值。

為什麼需要類別圖表？

對於類別資料，我們通常希望在圖表中視覺化地呈現某些統計值（例如平均值或計數）。雖然我們已經知道如何使用 pandas 獲取這些數據，但通過繪製圖表，我們可以更直觀地理解這

些數據。

類別圖表的應用這裡我們重點介紹如何在類別內進行統計估計。舉例來說，我們可以視覺化地報告某個統計值（如平均值或計數）在不同

````python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("dm_office_sales.csv")
df.head()
````

````python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x="division",data=df)
````

````python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x="level of education",data=df,hue="training level")
````

````python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x="level of education",data=df,hue="training level",palette="Set2")
````

Streamlit 是一個開源的 Python 庫，用於快速構建和共享數據應用。它設計簡單，讓你可以專注於編寫 Python 代碼，而無需擔心前端開發的複雜性。以下是 Streamlit 的一些主要特點和優勢：

### 主要特點

1. **即時重新運行**：
   - 每次你修改並保存 Python 腳本時，Streamlit 都會自動重新運行應用並更新結果，讓開發過程更加流暢。

2. **簡單易用的 API**：
   - Streamlit 提供了簡單而強大的 API，用戶可以通過簡單的函數調用來添加文本、數據表、圖表、交互控件等。

3. **互動控件**：
   - Streamlit 支持多種互動控件，如按鈕、滑塊、選擇框、文件上傳等，幫助用戶創建豐富的交互界面。

4. **支持多種數據可視化**：
   - Streamlit 支持與多種數據可視化庫（如 Matplotlib、Plotly、Altair 等）集成，可以輕鬆地展示各類圖表和數據視覺化結果。

5. **輕鬆部署和分享**：
   - 應用可以輕鬆部署到雲端，並通過鏈接與他人分享，讓團隊協作更加方便。

### 主要優勢

1. **降低開發成本**：
   - Streamlit 的設計使得開發者可以快速構建原型，減少了傳統前後端分離開發的複雜性和時間成本。

2. **即時反饋**：
   - 即時重新運行和更新功能讓開發者可以快速看到代碼變更的效果，提高了開發效率。

3. **靈活性和擴展性**：
   - Streamlit 可以輕鬆與其他 Python 庫集成，如數據處理的 Pandas、數據科學的 NumPy 和 SciPy，以及機器學習的 Scikit-learn 等，擁有很強的靈活性和擴展性。

### 安裝與入門

#### 安裝

你可以通過 pip 安裝 Streamlit：

```bash
pip install streamlit
```
運行以下測試
```bash
streamlit hello
```

#### 入門範例

以下是一個簡單的 Streamlit 應用範例：

```python
import streamlit as st

# 設置標題
st.title('Hello Streamlit')

# 獲取用戶輸入
user_input = st.text_input('Type something:')

# 顯示用戶輸入
st.write(f'You typed: {user_input}')
```

運行這段代碼：

```bash
streamlit run your_script.py
```

你的應用將在瀏覽器中打開，並顯示一個簡單的用戶界面。

### 常見應用場景

1. **數據科學儀表板**：
   - 展示數據分析結果、機器學習模型性能等。

2. **交互式報告**：
   - 創建動態數據報告，讓用戶可以自行篩選和探索數據。

3. **快速原型設計**：
   - 快速驗證概念和想法，展示給團隊和客戶。

Streamlit 讓 Python 開發者可以更輕鬆地構建專業的數據應用，並快速迭代和分享。希望這些簡介對你有所幫助！
